```python
import pandas as pd
from nameparser import HumanName
```

```python
df1 = pd.read_csv('dataset1.csv')
df2 = pd.read_csv('dataset2.csv')
```

```python
def split_name(name):
    n = HumanName(name)
    return pd.Series([n.first, n.last])
```

```python
# Remove rows which do not have a name
df1 = df1[df1['name'].notna()]
df2 = df2[df2['name'].notna()]
```

```python
# Split name into first_name and last_name
df1[['first_name', 'last_name']] = df1['name'].apply(lambda x: split_name(x))
df2[['first_name', 'last_name']] = df2['name'].apply(lambda x: split_name(x))
```

```python
# Remove 0 prepended to the price field. Can do this by converting the data type of 'price' column to float
df1['price'] = pd.to_numeric(df1['price'])
df2['price'] = pd.to_numeric(df2['price'])
```

```python
# Create new field if row has price > 100
df1['above_100'] = df1['price'].apply(lambda x: True if x>100 else False)
```
